# **Problem 3: Digital Image Acquisition**

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

# A.

```python
def generate_image(function, image_size, file_path):
    x, y = np.linspace(-1, 1, image_size[0]), np.linspace(-1, 1, image_size[1])
    X, Y = np.meshgrid(x, y)
    
    data = function(X, Y)
    data = np.clip(data, 0, 255).astype(np.uint8)  
    
    # Convert the array to an image and save it
    image = Image.fromarray(data)
    image.save(file_path)
    print(f"Image saved at: {file_path}")

def func_1(X, Y):
    return 255 * np.sin(20 * np.pi * np.sqrt(X**2 + Y**2))

def func_2(X, Y):
    result = np.sin(20 * np.pi * X) * np.sin(20 * np.pi * Y)
  #  np.where(condition, x, y)
    return np.where(result > 0, 255, 0)
```

# B.

```python
image_size = (512, 512)

generate_image(func_1, image_size, 'function_1.png')
generate_image(func_2, image_size, 'function_2.png')
```

# C.

```python
def generate_eq1(t):
    x = np.sin(t) * (np.exp(np.cos(t)) - 2 * np.cos(4*t) - np.sin(t/12)**5)
    y = np.cos(t) * (np.exp(np.cos(t)) - 2 * np.cos(4*t) - np.sin(t/12)**5)
    return x, y

# np.linspace(start, stop, num)
t_values = np.linspace(0, 12 * np.pi, 1000)

x1, y1 = generate_eq1(t_values)

plt.plot(x1, y1)
plt.title("Equation 1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.axis("equal")
plt.show()
```

```python
def  generate_eq2(theta, k=5):
    x = np.cos(k * theta) * np.cos(theta)
    y = np.cos(k * theta) * np.sin(theta)
    return x, y

theta_values = np.linspace(0, 2 * np.pi, 1000)

x2, y2 =  generate_eq2(theta_values)

plt.plot(x2, y2)
plt.title("Equation 2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.axis("equal")
plt.show()
```

# D.

```python
def generate_image_resolution(function, resolutions):
    for r in resolutions:
        # Points depend on the resolution
        t_values = np.linspace(0, 12 * np.pi, r)

        x1, y1 = function(t_values)

        plt.plot(x1, y1)
        plt.title(f"Equation 1 at Resolution {r}x{r}")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.grid(True)
        plt.axis("equal")
        plt.savefig(f'Equation 1_{r}x{r}.png')
        plt.close()

resolutions = [100, 500, 1000]
generate_image_resolution(generate_eq1, resolutions)
```

```python
def generate_image_resolution(function, resolutions, k=5):
    for r in resolutions:
        theta_values = np.linspace(0, 2 * np.pi, r)  
        
        x2, y2 = function(theta_values, k)

        plt.plot(x2, y2)
        plt.title(f"Equation 2 at Resolution {r}x{r}")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.grid(True)
        plt.axis("equal")
        plt.savefig(f'Equation 2_{r}x{r}.png')
        plt.close()

resolutions = [100, 500, 1000]
generate_image_resolution(generate_eq2, resolutions)
```
